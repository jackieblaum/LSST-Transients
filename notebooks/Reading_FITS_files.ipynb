```python
from astropy.io import fits as pyfits

import matplotlib.pyplot as plt

# This only if you are in a notebook
%matplotlib notebook

from matplotlib.colors import LogNorm 
```

```python
# If you just need one of the extensions
data = pyfits.getdata("lsst_e_220_f1_R22_S11_E000.fits.gz", 0)

# If you only need the header
header = pyfits.getheader("lsst_e_220_f1_R22_S11_E000.fits.gz", 0)

print(header['RA_DEG'])
```

```python
# If you need both

# The old way
f = pyfits.open("lsst_e_220_f1_R22_S11_E000.fits.gz")

data = f[0].data
header = f[0].header

f.close()

# The new way (always use this)
with pyfits.open("lsst_e_220_f1_R22_S11_E000.fits.gz") as f:
    
    data = f[0].data
    header = f[0].header
```
